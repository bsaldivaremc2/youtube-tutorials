```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np


cat_image = np.array(Image.open("Imgs/scream.jpg"))
f = plt.figure(figsize=(10,10),dpi=80)
plt.imshow(cat_image)

# Create a Rectangle patch
posx,posy = 100,300
width,height = 100,100
color="red"
plt.gca().add_patch(
          plt.Rectangle([posx,posy],
                      width,height,
                      fill=False,
                      color=color,
                      alpha=0.5)
          )
posx,posy = 200,200
width,height = 50,50
color="blue"
plt.gca().add_patch(
          plt.Rectangle([posx,posy],
                      width,height,
                      fill=True,
                      color=color,
                      alpha=0.5)
          )
plt.show()
```
